# txt to dmp

This python script will convert a txt list of image files into a "dmp" file, based on that found within constrained cnn's implementation.

### input should look like the following:

filename.jpg

filename2.jpg

filename3.jpg

.

.

.

filenameN.jpg

### output should look like the following:

(lp1

S'filename.jpg'

p2

aS'filename2.jpg'

p3

aS'filename3.jpg'

.

.

.

aS'filenameN.jpg'

pN

a.

```python
#Open and read from file, formatting text

file = open("datasetList.txt", 'r')

fileList = file.readlines()

output = "(lp1\nS"
count = 2


for i in fileList:
    name = i[0:-1]
    output += '\'' + name + '\'\np' + str(count) + '\naS'
    count += 1
    
output = output[0:-1] + "."
print(output)

file.close()
```

```python
#Load output into out file
outputFile = open("datasetOutput.dmp", 'w')

outputFile.write(output)

outputFile.close()
```
